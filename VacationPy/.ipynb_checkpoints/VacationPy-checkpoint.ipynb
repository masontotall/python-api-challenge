```python
#imports setup

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os
import json
import time


from api_keys import g_key
```

```python
#load csv from WeatherPy

weather_df = pd.read_csv("../WeatherPy/output_data/cities.csv")
weather_df
```

Configure Gmaps and plot the Heatmap

```python
# Configure gmaps

gmaps.configure(api_key=g_key)


# Store latitude and longitude in locations

locations = weather_df[["Lat", "Lng"]]


# Store Humidity in humidity

humidity = weather_df["Humidity"]
```

```python
# Plot Heatmap

g_map = gmaps.figure(center=(46.0, -5.0), zoom_level=2)
max_intensity = np.max(humidity)


# Create heat layer

heat_map = gmaps.heatmap_layer(locations, weights = humidity, dissipating=False, max_intensity=100, point_radius=3)


# Add layer

g_map.add_layer(heat_map)


# Display figure

g_map
```

Create Dataframe narrowed down to ideal weather conditions

```python
# Narrow dataframe to cities with max temps between 75F-90F, cloudiness of 0, and windspeed less than 10 mph

ideal_df = weather_df.loc[(weather_df["Max Temp"] >= 75) & (weather_df["Max Temp"] <= 90) & \
                                   (weather_df["Wind Speed"] <= 10) & (weather_df["Cloudiness"] == 0)].dropna()

ideal_df
```
